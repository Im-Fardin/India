In this exercise, we'll look at a dataset of Indian cuisine and examine how spicy Hindus are. We also find a few of their longest-cooked dishes.

```python
import numpy as np
import pandas as pd
```

```python
#first of all let's read the file
df = pd.read_csv('food.csv')
df.head()
```

We have heard a lot that Hindus love spicy food! Now, we want to check the validity of this proposition according to the existing dataset and see what proportion of the total foods in this collection have a spicy taste?
To do this, just calculate the result of dividing the number of spicy foods by the total number of foods and store them in <code>the spicy_ratio</code> variable.


```python
spicy_ratio = df['flavor_profile'].value_counts()[0] / np.sum(df['flavor_profile'].value_counts())
print(spicy_ratio)
```

Now we want to know which foods consume the most time from the moment you start preparing the side dishes to the moment they are ready to be served.

```python
df_most_time_consuming = df.copy()
df_most_time_consuming['total_prep_time']=df_most_time_consuming.apply(lambda df: df['prep_time'] + df['cook_time'] , axis=1)
df_most_time_consuming=df_most_time_consuming.sort_values(by=['total_prep_time','name'],ascending=[False,True],axis=0,ignore_index=True)
df_most_time_consuming=df_most_time_consuming.head(10)
df_most_time_consuming.drop(['total_prep_time'],axis=1 , inplace=True)
df_most_time_consuming
```
